```python
# Import
import matplotlib.pyplot as plt
```

plt.plot() → Creates a simple line plot

plt.show() → Displays the plot

**My First Plot: Line Plot**

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)
plt.show()
```

**Titles, Labels, and Grid**

```python
plt.plot(x, y)
plt.title("My First Line Chart")
plt.xlabel("X Axis - Time")
plt.ylabel("Y Axis - Sales")
plt.grid(True)
plt.show()
```

**Customize Style (Color, Line, Marker)**

```python
plt.plot(x, y, color='red', linestyle='--', marker='o')
plt.show()
```

**Multiple Lines on Same Graph**

```python
y2 = [5, 15, 20, 25, 30]

plt.plot(x, y, label='Product A')
plt.plot(x, y2, label='Product B')
plt.legend()  # Show labels
plt.show()
```

**Bar Chart**

```python
products = ['A', 'B', 'C']
sales = [100, 200, 150]

plt.bar(products, sales, color='skyblue')
plt.title("Product Sales")
plt.show()
```

**Pie Chart**

```python
sizes = [30, 45, 25]
labels = ['Apple', 'Banana', 'Mango']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Fruit Share")
plt.show()
```

**Scatter Plot**

```python
height = [150, 160, 165, 170, 175]
weight = [50, 55, 60, 70, 75]

plt.scatter(height, weight, color='purple')
plt.title("Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.grid(True)
plt.show()
```

**Histogram**

```python
import numpy as np

data = np.random.randn(1000)  # Random numbers
plt.hist(data, bins=20, color='green')
plt.title("Histogram Example")
plt.show()
```

**Subplots (Multiple Graphs in One Window)**

```python
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title("Line Chart")

plt.subplot(1, 2, 2)
plt.bar(products, sales)
plt.title("Bar Chart")

plt.tight_layout()  # Avoid overlap
plt.show()
```

```python
plt.plot(x, y)
plt.savefig("my_plot.png")  # Saves to file
```
